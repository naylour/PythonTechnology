```python
documents: list[dict[str, str]] = [
    {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
    {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
    {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"},
]

directories: dict[str, list[str]] = {
    "1": ["2207 876234", "11-2"],
    "2": ["10006"],
    "3": [],
}
```

```python
# Задание 1


def take_document_by_id(document_id: str) -> dict[str, str] | None:
    filtered_documents = list(
        filter(lambda document: document["number"] == document_id, documents)
    )

    if len(filtered_documents) != 1:
        return None

    return filtered_documents[0]


def take_directory_by_document_number(doc_number: str) -> str | None:
    for directory, doc_numbers in directories.items():
        if doc_number in doc_numbers:
            return directory

    return None


def getting_data_all_documents():
    return list(
        map(
            lambda document: f"№ {document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {take_directory_by_doc_number(document['number'])}",
            documents,
        )
    )


def create_directory(directory_number: str) -> bool:
    if directory_number in directories:
        return False

    directories[directory_number] = []

    return True


def drop_directory(directory_number: str) -> bool | None:
    current_directory = directories.get(directory_number)

    if current_directory is None:
        return None

    if len(current_directory) != 0:
        return False

    del directories[directory_number]

    return True
```

```python
# Задание 2


def create_document() -> bool:
    document_id = input("Введите номер документа:\n")
    document_type = input("Введите тип документа:\n")
    document_owner = input("Введите владельца документа:\n")
    document_directory = input("Введите полку для хранения:\n")

    if document_directory not in directories:
        return False

    documents.append({"type": document_type, "number": document_id, "name": document_owner})

    return True


def drop_document(document_id: str) -> bool:
    document = take_document_by_id(document_id)

    if document is None:
        return False

    documents[:] = [doc for doc in documents if doc['number'] != document['number']]

    return True


def move_document() -> bool | None:
    document_id = input("Введите номер документа:\n")
    directory = input("Введите номер полки:\n")

    if directory not in directories.keys():
        return None

    current_document = take_document_by_id(document_id)

    if current_document is None:
        return False

    maybe_directory = take_directory_by_document_number(document_id)

    if maybe_directory is not None:
        directories[maybe_directory] = list(filter(lambda doc: doc != current_document['number'], directories[maybe_directory]))

    directories[directory].append(document_id)

    return True
```

```python
while True:
    command = input('Введите команду:\n')

    if command == 'p':
        document_id = input('Введите номер документа:\n')
        document = take_document_by_id(document_id)

        if document is None:
            print('Документ не найден в базе')
        else:
            print(f'Владелец документа: {document['name']}')

    elif command == 's':
        document_id = input('Введите номер документа:\n')
        directory = take_directory_by_document_number(document_id)

        if directory is None:
            print('Документ не найден в базе')
        else:
            print(f'Документ хранится на полке: {directory}')

    elif command == 'l':
        documents_data = getting_data_all_documents()

        for document_data in documents_data:
            print(document_data)

    elif command == 'ads':
        directory_name = input('Введите номер полки:\n')

        is_created = create_directory(directory_name)

        if is_created:
            print('Полка добавлена. Текущий перечень полок: ', *directories.keys())
        else:
            print('Такая полка уже существует. Текущий перечень полок:', *directories.keys())

    elif command == 'ds':
        directory_name = input('Введите номер полки:\n')

        is_dropped = drop_directory(directory_name)

        if is_dropped is None:
            print('Такой полки не существует. Текущий перечень полок: ', *directories.keys())
        elif is_dropped:
            print('Полка удалена. Текущий перечень полок: ', *directories.keys())
        else:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', *directories.keys())

    elif command == 'ad':
        is_created = create_document()

        if is_created:
            print('Документ добавлен. ', end="")
        else:
            print('Такой полки не существует. Добавьте полку командой as.')

        print('Текущий список документов:')

        documents_data = getting_data_all_documents()

        for document_data in documents_data:
            print(document_data)

    elif command == 'd':
        document_id = input('Введите номер документа:\n')

        is_dropped = drop_document(document_id)

        if is_dropped:
            print('Документ удален.')
        else:
            print('Документ не найден в базе.')

        print('Текущий список документов:')

        documents_data = getting_data_all_documents()

        for document_data in documents_data:
            print(document_data)

    elif command == 'm':
        move_result = move_document()

        if move_result is None:
            print('Такой полки не существует. Текущий перечень полок: ', *directories.keys())

            continue
        elif move_result:
            print('Документ перемещён.')
        else:
            print('Документ не найден в базе.')

        print('Текущий список документов:')

        documents_data = getting_data_all_documents()

        for document_data in documents_data:
            print(document_data)

    elif command == 'q':
        break
```
